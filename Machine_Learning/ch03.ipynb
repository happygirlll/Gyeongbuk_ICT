# Seaborn 라이브러리

### 데이터 세터의 흐름을 그래프로 그리는 것에 최적화
- Seaborn의 다양한 그래프들
    - displot : 분포와 밀도를 확인할 수 있는 형태, 히스토그램과 커널 밀도 추정(KDE)을 그릴 수 있다. 데이터의 분포를 이해하는데 유용함
    - relplot : 2개의 열간의 상관관계를 나타내는 형태(= 두 변수 간의 관계를 시각화하기 위한 가장 일번적인 방법), 데이터의 관계를 보여주는 산점도(scatter plot)와 선 플롯(line plot)을 쉽게 생성할 수 있다. 다양한 차원으로 데이터를 표현할 수 있으며, 데이터의 여러 범주를 색상, 모양, 크기 등으로 시각화하는데 유용함. 기본적으로 relplot은 산점도를 생성함.

    - pairplot : 그리드형태로 각 집합의 조합에 대해 히스토그램과 분포도를 나타내는 형태, 데이터셋의 여러 변수 간의 쌍별 관계를 시각화하는데 사용. 각 변수 쌍에 대해 산점도를 그리고, 대각선에는 각 변수의 히스토그램(또는 KDE)을 표시. 범주형 데이터에 따라 색상을 다르게 지정할 수 있어, 데이터의 다양한 범주를 시각적으로 구분하기에 유용함.
    - countplot : 항목별 갯수를 나타내는 막대그래프 형태
    - boxplot : 데이터 분포와 주요 통계량을 시각화하는 박스 형태
    - swarmplot : 범주형 데이터의 분포를 시각화하는 형태
    - violinplot : 데이터의 분포와 밀도는 시각화하는 형태


```python
import seaborn as sns

#데이터셋 로드 : 'seaborn'에서 제공하는 'tips' 데이터셋을 로드함
#tips 데이터셋은 식당 팁에 대한 정보를 담고 있음
tips_dataset = sns.load_dataset('tips')
tips_dataset.head()
```

```python
# 히스토그램 생성
sns.displot(tips_dataset['total_bill'], bins=30, kde=False)
#total_bill 열에 대해 히스토그램을 생성함
#bins=30 : 히스토그램을 30개의 구간으로 나눔
#kde=False : 커널 밀도 추정을 비활성화함 (기본적으로 true)
```

```python
#seaborn의 내장 데이터셋인 tips를 사용하여, 총 청구 금액(total_bill)과 팁(tip) 사이의 관계를 시각화
sns.relplot(x='total_bill', y='tip', data=tips_dataset)
```

```python
# relplot 추가옵션
'''
hue : 다른 범주에 따라 데이터 포인트의 색상을 다르게 설정 가능
style : 다른 범주에 따라 데이터 포인트의 선 스타일을 다르게 설정 가능
size : 다른 범주에 따라 데이터 포인트의 크기를 다르게 설정 가능
'''
sns.relplot(x='total_bill', y='tip', hue='sex', data=tips_dataset)
```

```python
# tips 데이터셋을 사용하여 성별(sex)에 따라 색상을 다르게 지정하고 쌍별 플롯을 생성하는 예제
sns.pairplot(tips_dataset, hue='sex', palette='coolwarm')
# hue='sex'를 사용하여 'sex'의 변수에 따라 포인트의 색상을 다르게 설정
# 'palette='coolwarm'을 사용하여 색상 팔레트를 적용
```

```python
# pairplot의 추가옵션
'''
dig_kind='kde' : 대각선의 히스토그램 대신 KDE를 사용
diag_kws={'fill':False} : 대각선의 히스토그램을 채우지 않음
markers : 매개변수를 사용하여 산점도의 마커 스타일을 변경할 수 있음
변수선택 : 'vars' 매개변수를 사용해 시각화할 변수의 하위 집합을 지정
'''
sns.pairplot(tips_dataset, hue='sex', palette='coolwarm', diag_kind='reg')
```
