```python
#df.loc : 라벨 기반 인덱싱을 사용하여 데이터프레임의 특정 행이나 열을 선택
'''
 [주요특징]
 - 라벨 기반 : 인덱스와 열의 이름을 사용하여 데이터를 선택
 - 행과 열을 모두 인덱싱 가능 : 'df.loc[row_label, col_label]' 형식으로 사용하여 특정 행과 열을 지정 가능
 - 슬라이싱 포함 : 슬라이싱할 때 종료 라벨도 포함됨

 '''
Arr_df.loc['E']
```

```python
# df.iloc : 정수기반 인덱싱을 사용하여 데이터프레임의 특정 행이나 열을 선택. 즉 데이터의 정수 위치를 기반으로 선택함
'''
[주요 특징]
- 정수 기반 : 정수 인덱스를 사용하여 데이터를 선택
- 0부터 시작하는 정수 인덱스 사용 : 'df.iloc[row_index, col_index]' 형식으로 사용하여 특정 행과 열을 지정 가능
- 슬라이싱 포함 : 슬라이싱할 때 종료 인덱스도 포함되지 않는다.
'''
Arr_df.iloc[4]
```
